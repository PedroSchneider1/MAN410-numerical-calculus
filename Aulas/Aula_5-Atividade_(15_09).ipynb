```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""
a) calcular o coeficiente angular a ou a1 da reta. Usar duas casas decimais.
b) calcular o coeficiente linear b ou a0 da reta. Usar duas casas decimais.
c) Calcular o resíduo do ajuste. Usar duas casas decimais.
d) determinar o valor de y para x=2. Usar duas casas decimais.
e) determinar o coeficiente de correlação linear. Usar três casas decimais.
f) calcular o coeficiente de determinação R2. Usar três casas decimais.
"""
# 1

x = [4.04, 4.7, 5.39, 6, 6.5]
y = [0.586, 0.358, 0.292, 0.243, 0.225]

x_np = np.array(x)
y_np = np.array(y)

n = np.size(x_np)
Sx= np.sum(x_np)
Sy=np.sum(y_np)
Sxy=np.sum(x_np*y_np)
Sxx=np.sum(x_np*x_np)
a1=(n*Sxy-Sx*Sy)/(n*Sxx-Sx**2)   #Calcula o coeficiente angular a1 da reta
a0=(Sxx*Sy-Sxy*Sx)/(n*Sxx-Sx**2) #Calcula o coeficiente linear a0 da reta

print(f"(a) a1: {a1:.2f}")
print(f"(b) a0: {a0:5.2f}")
```

```python
def Residuo(x,y,b0,b1):
    n = len(y)
    RS = 0
    for i in range(0,n):
         y_pred=b0+b1*x[i]
         RS = RS + (y[i]-y_pred)**2
    return RS
print(f"(c) RS: {Residuo(x_np,y_np,a0,a1):.2f}")
```

```python
x = 2
# f(x)=a1x+a0
y_pred = a1*x + a0

# mostra a reta de regressão
plt.scatter(x, y_pred, color = "r")

plt.xlabel('x', fontsize = 15)
plt.ylabel('y', fontsize = 15)
plt.show(True)

print(f"(d) {y_pred:.2f}")
```

```python
coef = np.corrcoef(x_np, y_np)
print(f"(e) coeficiente: {coef[0][1]:.3f}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

x = x_np.reshape(-1, 1)
modelo = LinearRegression().fit(x, y)
R2 = r2_score(y, modelo.predict(x))
print(f'(f) R2 Score: {R2:.3f}')
```

```python
# 2
plt.scatter(x, y, color = "b", marker = "o", s = 50)

# prediz os valores
y_pred = a0 + a1*x_np

# mostra a reta de regressão
plt.plot(x, y_pred, color = "r")

plt.xlabel('x', fontsize = 15)
plt.ylabel('y', fontsize = 15)
plt.show(True)
```
