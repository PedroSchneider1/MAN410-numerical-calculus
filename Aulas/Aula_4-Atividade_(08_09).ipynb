```python
# 1)
# Utilizando localização gráfica ou o teorema de Bolzano, determinar o intervalo que contém  a menor raiz da equação
# f(x) = x^2+0.5x-0.5-cosx=0.
# Utilizar a amplitude de uma unidade para o intervalo da raiz na resposta. Exemplo ]8:9[ ; ]-5:-4[

x = np.linspace(-4,4,9)
y1 = x**2+0.5*x-0.5
y2=np.cos(x)

plt.plot(x, y1, label='x² + 0.5x - 0.5')
plt.plot(x, y2, label='cos(x)')
plt.grid(True)
plt.legend()
plt.show()

f = x**2+0.5*x-0.5-np.cos(x)
x = np.array([i for i in range(-4,5,1)]).reshape((1,9))
print(f)
```

```python
# 2)
# Calcular a maior raiz pelo método de Newton Rapshon ou pela  função fsolve da seguinte equação:
# x^2+sen(x)-5=0

x=sp.Symbol('x')
def NewtonsMethod(x0):
    tolerance=0.000001
    d=sp.diff(f(x),x)
    while True:
        x1 = x0 - (f(x0)/d.doit().subs({x:x0}))
        t = abs(x1 - x0)
        if t < tolerance:
            break
        x0 = x1
    return x1

def f(x):
    return x**2+sp.sin(x)-5

a = 2
b = 3
print("Ponto a: ", f(a).evalf())
print("Ponto b: ", f(b).evalf())

xi = (a + b) / 2  # ponto médio = 2.5
xr = NewtonsMethod(xi)
print('Ponto médio: ', xi)
print('Raiz: ', xr)
```
